```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import bz2
import pickle
import os.path
from collections import Counter
from pprint import pprint
import random
import os, PIL
from PIL import Image
import gzip
```

```python
#load all images and save to pickle file
database = []
print('test')
for root, dirs, files in os.walk('solved/'):
    for fname in files:
        if fname.endswith('.jpg') or fname.endswith('.png'):
            if any(elem in list('0-vj9c')  for elem in fname.split('_')[0]):
                print(fname)
                continue
            p = os.path.join(root,fname)
            label = list(fname.split('_')[0])
            image = cv2.imread(p)
            if fname.endswith('.jpg'):
                os.rename(os.path.join(root,fname),os.path.join(root,fname[:-4]+'.png'))
            database.append([label, image])
            
```

```python
with bz2.open('database.pkl.bz2', 'wb') as f:
    pickle.dump(database, f)
```

```python
#visualize labels
all_labels_chars = []
for label, picture in database:
    all_labels_chars.extend(label)

D = dict(Counter(all_labels_chars))
#plt.bar(all_labels_chars)
plt.bar(range(len(D)), list(D.values()), align='center')
plt.xticks(range(len(D)), list(D.keys()))

plt.show()

#all_labels_chars = set(all_labels_chars)
all_letters = list(set(all_labels_chars))

pprint(D)

print('all_letters:',all_letters)
```

```python
#show some pictures
picts = random.sample(database,3)
for p in picts:
    plt.title(''.join(p[0]))
    plt.imshow(p[1])
    plt.show()
```

```python
#preprocess data
#average
avg=np.zeros(database[0][1].shape,np.float)
N = len(database)

# Build up average pixel intensities, casting each image as an array of floats
for l, im in database:
    imarr=np.array(im,dtype=np.float)
    avg=avg+imarr

# Round values in array and cast as 8-bit integer
avg=np.array(avg/N,dtype=np.uint8)

# Generate, save and preview final image
out=Image.fromarray(avg,mode="RGB")
out.save("../Average.png")
plt.title('Average')
plt.imshow(out)
plt.show()
```

```python
#subtract avg from all images
database_sub = []
for i in range(N):
    database_sub.append([database[i][0],database[i][1]-avg])

#show some pictures
picts = random.sample(database_sub,3)
for p in picts:
    plt.title(''.join(p[0]))
    plt.imshow(p[1])
    plt.show()
```

```python
#save dataset
with bz2.open('database_sub.pkl.bz2', 'wb') as f:
    pickle.dump(database_sub, f)
```

```python
BLOCK_SIZE = 1000
THRESHOLD = 25

def preprocess(image):
    image = cv2.medianBlur(image, 5)
    image = cv2.GaussianBlur(image, (5, 5), 0)
    return image


def postprocess(image):
    image = cv2.medianBlur(image, 3)
    # image = cv2.medianBlur(image, 5)
    kernel = np.ones((2,2), np.uint8)
    image = cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
    return image


def get_block_index(image_shape, yx, block_size): 
    y = np.arange(max(0, yx[0]-block_size), min(image_shape[0], yx[0]+block_size))
    x = np.arange(max(0, yx[1]-block_size), min(image_shape[1], yx[1]+block_size))
    return np.meshgrid(y, x)


def adaptive_median_threshold(img_in):
#     med = np.median(img_in)
#     img_out = np.zeros_like(img_in)
#     img_out[img_in - med < THRESHOLD] = 255
    return img_in


def block_image_process(image, block_size):
    out_image = np.zeros_like(image)
    for row in range(0, image.shape[0], block_size):
        for col in range(0, image.shape[1], block_size):
            idx = (row, col)
            block_idx = get_block_index(image.shape, idx, block_size)
#             out_image[block_idx] = adaptive_median_threshold(image[block_idx])
            out_image[block_idx] = image[block_idx]

    return out_image


def process_image_file(image_in):
    if len(image_in.shape) == 3:
        if image_in.shape[2] == 3:
            image_in = cv2.cvtColor(image_in, cv2.COLOR_BGR2GRAY)

    image_in = preprocess(image_in)
    image_out = block_image_process(image_in, BLOCK_SIZE)
    image_out = postprocess(image_out)

    return image_out

#show some pictures
picts = random.sample(database,3)
for p in picts:
    plt.title(''.join(p[0]))
    plt.imshow(Image.fromarray(process_image_file(p[1])))
    plt.figure()
    plt.imshow(Image.fromarray(p[1]))
    plt.show()

```

```python
database_sub2 = []
for i in range(N):
    database_sub2.append([database[i][0],process_image_file(database[i][1])])
```

```python
def crop(p):
    sx = np.sum(p,axis=0)
    sy = np.sum(p,axis=1)
    a = np.argmax(sx[10:]>3500)+10
    b = np.argmax(sx[120:]<3200)+120
    c = np.argmax(sy[4:]>6000)+4
    d = np.argmax(sy[70:]<6000)+70
    if a == 10:
        a = 0
    if b == 120:
        b = len(sx)
#     print(p.shape)
#     print(a,b,c,d)
#     c = 0
#     d = 80
    return cv2.resize(p[c:d,a:b],(180,80))
    
database_sub5 = []
for i in range(N):
    database_sub5.append([database_sub4[i][0],crop(database_sub4[i][1])])
    
#show some pictures
picts = random.sample(database_sub5,3)
for p in picts:
    plt.title(''.join(p[0]))
    plt.imshow(Image.fromarray(p[1]))
    plt.show()
#save dataset
with bz2.open('../database_sub5.pkl.bz2', 'wb') as f:
    pickle.dump(database_sub5, f)
```
